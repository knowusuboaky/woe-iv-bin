```python
## Load Package
from woe_iv_bin import categorical_woe
from woe_iv_bin import continuous_woe
```

```python
import pandas as pd
import numpy as np
```

```python
## Data

# Seed for reproducibility
np.random.seed(42)

# Generating synthetic data
n = 1000  # Number of samples
df = pd.DataFrame({
    'target': np.random.binomial(1, p=0.3, size=n),  # Binary target variable with a 30% chance of being 1
    'cat_feature': np.random.choice(['Category 1', 'Category 2', 'Category 3'], size=n),  # Categorical feature
    'cont_feature': np.random.normal(loc=50, scale=15, size=n)  # Continuous feature
})

# Display the first few rows of the dataframe
df.head()
```

```python
# Categorical 
woe_results = categorical_woe(df, cat_variable_name='cat_feature', y_df=df['target'])
```

```python
woe_results
```

```python
## Continuous 
df_binned, optimized_bins, binning_woe_results = continuous_woe(df, feature = 'cont_feature', target = 'target', max_bins=20, min_samples_bin=0.05)
```
